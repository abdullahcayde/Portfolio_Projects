```python
# Recap 01
# ---------------------------- Libraries ----------------------------
import time
from datetime import datetime

import pandas as pd
import numpy as np

#pd.options.display.max_colwidth = 50

import warnings
warnings.simplefilter(action='ignore')

# ---------------------------- Functions ----------------------------
def replace_column(data, column_name, old, new):
    data[column_name] = data[column_name].str.replace(old, new)

# ---------------------------- DataFrames ----------------------------
path = '/Users/macbook/Desktop/projects/Github_Repositories/Portfolio Projects/02 - Web_Scraping_Job_Search/data'

time_ = datetime.today().strftime('%Y-%m-%d')
print(time_)
job_name = 'Data Analyst'
job_name2 = job_name.replace(' ', '-')
jobs = pd.read_csv(f'{path}/{job_name2}-{time_}.csv')
df1 = jobs.copy()


job_name = 'Business Analyst'
job_name2 = job_name.replace(' ', '-')
df2 = pd.read_csv(f'{path}/{job_name2}-{time_}.csv')


job_name = 'Data Scientist'
job_name2 = job_name.replace(' ', '-')
df3 = pd.read_csv(f'{path}/{job_name2}-{time_}.csv')


# ---------------------------- Concatinate DataFrames ----------------------------
df = pd.concat([df1,df2,df3], axis=0, ignore_index=True)
print(f'{df1.shape}, {df2.shape}, {df3.shape}, {df.shape}')

# ---------------------------- Cleaning ----------------------------
df.replace({'job_title' : {'\(m/w/d\)':'', '\(m/f/d\)':'', '\(w/m/x\)':'', '\(w/m/d\)':'', '\(f/m/x\)':'' }}, regex=True, inplace=True)
df.company = df.company.str.lower()
df.job_title = df.job_title.str.lower()
df.city = df.city.str.lower()

first = df.shape[0]
print(f'Shape first : {df.shape}')
# ---------------------------- Drop Company Null Values ----------------------------
df.dropna(subset=['company'], inplace=True)

# ---------------------------- Drop Company Null Values ----------------------------
df.drop_duplicates(subset=['job_title', 'company', 'city', 'description'], inplace=True)
last = df.shape[0]
print(f'Shape Last : {df.shape}, Dropped : {first - last}')

df.head()
```

# Filtering

```python
df[df.job_title.str.contains('data', regex=True)].job_title.value_counts().reset_index().sort_values('job_title', ascending=False).head(10)
```

```python
pd.options.display.max_colwidth = 70
df.loc[df.job_title.str.contains('werkstudent')]
```

```python
# No Filter in job title
df.company.value_counts().reset_index().sort_values('company', ascending=False).head(10)
```

```python
pd.options.display.max_colwidth = 70
df.loc[df.job_title.str.contains('data analyst|data scientist|business analyst')].groupby('company').agg({'job_title' : ['count', 'unique'],
                                                                                         'city' : ['unique']}).sort_values([('job_title','count')], ascending=False).head(15)
```

```python
# Filter 'Data' in job title
df[df.job_title.str.contains('data analyst|data scientist|business analyst')].company.value_counts().reset_index().sort_values('company', ascending=False).head(30)
```

```python
# # Filter 'Data|Business|BI|Intelligence' in job title
df[df.job_title.str.contains('data|business|bi|intelligence|scientist')].company.value_counts().reset_index().sort_values('company', ascending=False).head(10)
```

```python
# # Filter  include 'Data|Business|BI|Intelligence' and  exclude 'SAP|Engineer|Cloud|Senior|Warehouse|Lead'  in job title
dff = df.loc[(df.job_title.str.contains('data|business|bi|intelligence|scientist', regex=True)) & (~df.job_title.str.contains('sap|engineer|cloud|senior|warehouse|lead', regex=True))]
dff.company.value_counts().reset_index().sort_values('company', ascending=False).head(10)
```
