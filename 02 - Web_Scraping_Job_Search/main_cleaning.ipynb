```python
# Recap 01
from datetime import timedelta

def replace_column(data, column_name, old, new):
    data[column_name] = data[column_name].str.replace(old, new)

job_name = 'Data Analyst'
path = '/Users/macbook/Desktop/projects/Github_Repositories/Portfolio Projects/02 - Web_Scraping_Job_Search/data'
job_name2 = job_name.replace(' ', '-')
time_ = datetime.today().strftime('%Y-%m-%d')
jobs = pd.read_csv(f'{path}/{job_name2}-{time_}.csv')

df = jobs.copy()

print(f'Shape first : {df.shape}')

df.dropna(subset=['company'], inplace=True)

df.drop_duplicates(subset=['job_title', 'company', 'city'], inplace=True)
print(f'Shape Last : {df.shape}')

# Recap 02
# Fill NAN to no_info
df.publish = df.publish.str.lower()
df.publish.fillna('no_info', inplace=True)

df.replace({'job_title' : {'\(m/w/d\)':'', '\(m/f/d\)':'', '\(w/m/x\)':'', '\(w/m/d\)':'', '\(f/m/x\)':'' }}, regex=True, inplace=True)

df.head()
```

```python
df[df.job_title.str.contains('Data')].job_title.value_counts().reset_index().sort_values('job_title', ascending=False).head(10)
```

```python
# No Filter in job title
df.company.value_counts().reset_index().sort_values('company', ascending=False).head(10)
```

```python
# Filter 'Data' in job title
df[df.job_title.str.contains('Data')].company.value_counts().reset_index().sort_values('company', ascending=False).head(10)
```

```python
# # Filter 'Data|Business|BI|Intelligence' in job title
df[df.job_title.str.contains('Data|Business|BI|Intelligence')].company.value_counts().reset_index().sort_values('company', ascending=False).head(10)
```

```python
# # Filter  include 'Data|Business|BI|Intelligence' and  exclude 'SAP|Engineer|Cloud|Senior|Warehouse|Lead'  in job title
dff = df.loc[(df.job_title.str.contains('Data|Business|BI|Scientist', regex=True)) & (~df.job_title.str.contains('SAP|Engineer|Cloud|Senior|Warehouse|Lead', regex=True))]
dff.company.value_counts().reset_index().sort_values('company', ascending=False).head(10)
```
